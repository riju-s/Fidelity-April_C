```python
import pandas as pd
```

```python
staff = pd.DataFrame({"Staff" : ["Mary", "John", "Peter"],
                      "City": ["London", "Paris", "Madrid"]})
```

```python
staff
```

```python
salaries = pd.DataFrame({"Cities": ["London", "Paris", "Sydney"],
                          "AvgSalary" : [10000, 20000, 30000]})
```

```python
pd.concat([staff, salaries])
```

```python
staff.join(salaries, lsuffix="_st", rsuffix="_sa")
```

```python
staff.set_index('City').join(salaries.set_index('Cities'))
```

```python
staff.join(salaries.set_index('Cities'), on='City')
```

```python
#
```

```python
pd.merge(staff, salaries, how="inner")
```

```python
pd.merge(staff, salaries, how="left")
```

```python
pd.merge(staff, salaries, how="right")
```

```python
pd.merge(staff, salaries, how="outer", left_on="City", right_on="City")
```
