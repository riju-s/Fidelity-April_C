```python
a=[1,2,3,4,5]
b=[2,4,6,8,10]
```

```python
fig,ax = plt.subplots()
ax.plot(a,b, marker="d", color="crimson", linestyle="dotted")
ax.plot(b,a)
ax.bar(a,b, color="slategray")
ax.scatter(b,a)
plt.show()
```
