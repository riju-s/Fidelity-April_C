```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"][0:11], 
        seattle_weather["MLY-TAVG-NORMAL"][0:11],
        marker="v", linestyle="--", color='r')
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average Temperature (F)")
ax.set_title("Weather in Seattle")
plt.show()
```

```python
fig, ax = plt.subplots(2,3, sharey=True, sharex=True)
ax[0,0].plot(seattle_weather["DATE"][0:11], 
        seattle_weather["MLY-PRCP-NORMAL"][0:11],
        marker="o",  color='r')
ax[0,0].plot(seattle_weather["DATE"][0:11], 
        seattle_weather["MLY-PRCP-25PCTL"][0:11],
        marker="o",  linestyle="--", color='r')
ax[0,0].plot(seattle_weather["DATE"][0:11], 
        seattle_weather["MLY-PRCP-75PCTL"][0:11],
        marker="o",  linestyle="--", color='r')
ax[1,0].set_xlabel("Time (months)")
ax[0,0].set_ylabel("Precipitation (in)")
ax[0,0].set_title("Weather in Seattle")
plt.show()
```
